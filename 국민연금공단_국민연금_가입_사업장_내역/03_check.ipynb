```python
import os
from glob import glob

import pandas as pd
```

```python
addr_paths = glob("./data/preproc_data/addr_*.csv")
subscribe_paths = glob("./data/preproc_data/subscribe_*.csv")
category_paths = glob("./data/preproc_data/category_*.csv")
```

```python
addr_df = pd.concat([pd.read_csv(file_path, index_col=0) for file_path in addr_paths], axis=0)
display(addr_df.sample(5))
addr_df.to_csv("./data/final_data/addr.csv")
```

```python
subscribe_df = pd.concat(
    [pd.read_csv(file_path, index_col=0) for file_path in subscribe_paths], axis=0
)
display(subscribe_df.sample(5))
subscribe_df.to_csv("./data/final_data/subscribe.csv")
```

```python
category_df = pd.concat(
    [pd.read_csv(file_path, index_col=0, low_memory=False) for file_path in category_paths], axis=0,
)
display(category_df.sample(5))
category_df.to_csv("./data/final_data/category.csv")
```
