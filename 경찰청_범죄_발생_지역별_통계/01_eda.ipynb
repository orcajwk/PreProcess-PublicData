```python
column_dict = dict()

for file_path in file_paths:
    df = pd.read_csv(file_path, encoding="cp949")
    column_dict[file_path] = df.columns
```

```python
columns_list = list(column_dict.values())
columns_list = [list(columns) for columns in columns_list]
```
