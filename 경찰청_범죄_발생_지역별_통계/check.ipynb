```python
from glob import glob
import pandas as pd
```

```python
file_paths = sorted(glob("./data/preproc_data/*.csv"))
```

```python
read_csv = lambda file_path: pd.read_csv(file_path, index_col=0)
```

```python
crime_df = pd.concat([read_csv(file_path) for file_path in file_paths], axis=0)
```

```python
display(crime_df.sample(5))
display(crime_df[(crime_df["지역"] == "서울") & (crime_df["범죄중분류"] == "마약범죄")])
```
